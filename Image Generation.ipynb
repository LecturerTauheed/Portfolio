```python
from diffusers import DiffusionPipeline
import torch
pipe = DiffusionPipeline.from_pretrained(
    #model
    "stabilityai/stable-diffusion-2-1-base",
    #dtype
    torch_dtype=torch.float16,
    #safe tensors
    use_safetensors=True,
    # variant
    variant="fp16",

    )
pipe.to("cuda")
```

```python
prompt= "Create me Pakistani X Prime minister Imran Khan Pic in standard suit"
negative_prompt = "No one available with him only imran khan"
image = pipe(
    prompt=prompt,
    negative_prompt=negative_prompt).images[0]
```

```python
plt.imshow(image)
plt.axis("off")
plt.show()
```

```python
def generate_image(prompt, negative_prompt):
  images = pipe(prompt=prompt, negative_prompt=negative_prompt).images[0]
  plt.imshow(images)
  plt.axis("off")
  plt.show()
```

```python
prompt= "A portrait of the famous cricketer Mahindra sing dohni in cricket uniform"
negative_prompt = "No one available with him only Virat Kohli"
generate_image(prompt, negative_prompt)
```
